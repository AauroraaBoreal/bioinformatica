```python
import pandas as pd
import numpy as np
import requests
import time
import re
import os
import gzip
import shutil
import urllib.request
import zipfile
from Bio import Entrez
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import classification_report

print("Librerías importadas correctamente")
```

```python
Entrez.email = "email@usil.pe"

url = "https://ftp.ncbi.nlm.nih.gov/pub/clinvar/vcf_GRCh38/clinvar.vcf.gz"
print("Descargando VCF de ClinVar...")
urllib.request.urlretrieve(url, "clinvar.vcf.gz")

with gzip.open("clinvar.vcf.gz", "rb") as f_in:
    with open("clinvar.vcf", "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)
print("Listo")
```

```python
print("Leyendo VCF...")
filas = []
with open("clinvar.vcf", "r") as f:
    for linea in f:
        if linea.startswith("#"):
            continue
        partes = linea.strip().split("\t")
        if len(partes) < 8:
            continue
        info = partes[7]
        if "BRCA1" not in info:
            continue
        clnsig = ""
        for campo in info.split(";"):
            if campo.startswith("CLNSIG="):
                clnsig = campo.replace("CLNSIG=", "")
                break
        filas.append({
            "chr": partes[0], "pos": partes[1],
            "ref": partes[3], "alt": partes[4],
            "clnsig": clnsig
        })

df_brca1 = pd.DataFrame(filas)
print("Variantes BRCA1:", len(df_brca1))
print(df_brca1["clnsig"].value_counts())
```

```python
def etiquetar(clnsig):
    c = clnsig.lower()
    if "pathogenic" in c and "benign" not in c and "conflicting" not in c:
        return 1
    elif "benign" in c and "pathogenic" not in c and "conflicting" not in c:
        return 0
    elif "uncertain" in c:
        return "vus"
    return None

df_brca1["etiqueta"] = df_brca1["clnsig"].apply(etiquetar)

df_entrenamiento = df_brca1[df_brca1["etiqueta"].isin([0, 1])].copy()
df_entrenamiento["etiqueta"] = df_entrenamiento["etiqueta"].astype(int)
df_vus = df_brca1[df_brca1["etiqueta"] == "vus"].copy()

df_entrenamiento = df_entrenamiento[
    (df_entrenamiento["ref"].str.len() == 1) &
    (df_entrenamiento["alt"].str.len() == 1)
].reset_index(drop=True)

df_vus = df_vus[
    (df_vus["ref"].str.len() == 1) &
    (df_vus["alt"].str.len() == 1)
].reset_index(drop=True)

print("Entrenamiento:", len(df_entrenamiento))
print("  Patogénicas (1):", (df_entrenamiento["etiqueta"]==1).sum())
print("  Benignas    (0):", (df_entrenamiento["etiqueta"]==0).sum())
print("VUS:", len(df_vus))
```

```python
# Cargamos el archivo CADD que ya descargamos con tabix
# Las columnas son: chr, pos, ref, alt, RawScore, PHRED
df_cadd = pd.read_csv(
    "cadd_brca1.tsv",
    sep="\t",
    header=None,
    names=["chr", "pos", "ref", "alt", "raw_score", "cadd_phred"]
)
df_cadd["pos"] = df_cadd["pos"].astype(str)
df_cadd["chr"] = df_cadd["chr"].astype(str)

print("Filas CADD:", len(df_cadd))
print(df_cadd.head())
```

```python
print("Cargando REVEL chr17...")
chunks = []
for chunk in pd.read_csv(
    "revel_data/revel_with_transcript_ids",
    sep=",", low_memory=False, chunksize=100000
):
    filtrado = chunk[chunk.iloc[:, 0].astype(str) == "17"]
    if len(filtrado) > 0:
        chunks.append(filtrado)

df_revel = pd.concat(chunks, ignore_index=True)
df_revel["grch38_pos"] = df_revel["grch38_pos"].astype(str)
df_revel["chr"] = df_revel["chr"].astype(str)
print("Filas REVEL:", len(df_revel))
print(df_revel.head())
```

```python
df_entrenamiento["pos"] = df_entrenamiento["pos"].astype(str)
df_entrenamiento["chr"] = df_entrenamiento["chr"].astype(str)

# Cruce con CADD
df_con_cadd = df_entrenamiento.merge(
    df_cadd[["chr", "pos", "ref", "alt", "cadd_phred"]],
    on=["chr", "pos", "ref", "alt"],
    how="left"
)

# Cruce con REVEL
df_con_todo = df_con_cadd.merge(
    df_revel[["grch38_pos", "ref", "alt", "REVEL"]],
    left_on=["pos", "ref", "alt"],
    right_on=["grch38_pos", "ref", "alt"],
    how="left"
).drop(columns=["grch38_pos"])

print("Con CADD:", df_con_todo["cadd_phred"].notna().sum())
print("Con REVEL:", df_con_todo["REVEL"].notna().sum())
print("Con ambos:", df_con_todo[["cadd_phred","REVEL"]].notna().all(axis=1).sum())
print()
print(df_con_todo[["chr","pos","ref","alt","cadd_phred","REVEL","etiqueta"]].head(10))
```

```python
def consultar_gnomad_coords(chr_, pos, ref, alt):
    variante_id = f"{chr_}-{pos}-{ref}-{alt}"
    query = """
    query {
      variant(variantId: "%s", dataset: gnomad_r4) {
        genome { af }
        exome  { af }
      }
    }
    """ % variante_id
    try:
        response = requests.post(
            "https://gnomad.broadinstitute.org/api",
            json={"query": query}, timeout=30
        )
        if response.status_code != 200:
            return 0.0
        data = response.json().get("data", {}).get("variant", None)
        if data is None:
            return 0.0
        genome_af = (data.get("genome") or {}).get("af", None)
        exome_af  = (data.get("exome")  or {}).get("af", None)
        if genome_af is not None:
            return genome_af
        if exome_af is not None:
            return exome_af
        return 0.0
    except Exception:
        return 0.0

# Solo usamos variantes que tengan CADD y REVEL
df_modelo = df_con_todo[
    df_con_todo["cadd_phred"].notna() &
    df_con_todo["REVEL"].notna()
].copy().reset_index(drop=True)

print(f"Consultando gnomAD para {len(df_modelo)} variantes...")
afs = []
for i, fila in df_modelo.iterrows():
    af = consultar_gnomad_coords(fila["chr"], fila["pos"], fila["ref"], fila["alt"])
    afs.append(af)
    if (i + 1) % 100 == 0:
        print(f"  {i+1}/{len(df_modelo)} procesadas")
    time.sleep(0.3)

df_modelo["af"] = afs
df_modelo.to_csv("brca1_dataset_final.csv", index=False)

print("\nDataset final:")
print(f"  Total: {len(df_modelo)}")
print(f"  Patogénicas (1): {(df_modelo['etiqueta']==1).sum()}")
print(f"  Benignas    (0): {(df_modelo['etiqueta']==0).sum()}")
print()
print(df_modelo[["chr","pos","ref","alt","cadd_phred","REVEL","af","etiqueta"]].head())
```

```python
FEATURES = ["cadd_phred", "REVEL", "af"]
X = df_modelo[FEATURES].values
y = df_modelo["etiqueta"].values

print("Distribución de clases:")
print(f"  Patogénicas (1): {(y==1).sum()} ({(y==1).mean()*100:.1f}%)")
print(f"  Benignas    (0): {(y==0).sum()} ({(y==0).mean()*100:.1f}%)")

modelo = RandomForestClassifier(
    n_estimators=200,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv = cross_validate(
    modelo, X, y, cv=skf,
    scoring=["precision", "recall", "f1"]
)

print("\nResultados validación cruzada 5 pliegues:")
print(f"  Precisión   : {cv['test_precision'].mean():.4f} ± {cv['test_precision'].std():.4f}")
print(f"  Sensibilidad: {cv['test_recall'].mean():.4f} ± {cv['test_recall'].std():.4f}")
print(f"  F1-score    : {cv['test_f1'].mean():.4f} ± {cv['test_f1'].std():.4f}")

modelo.fit(X, y)

print("\nImportancia de features:")
for feat, imp in zip(FEATURES, modelo.feature_importances_):
    barra = "█" * int(imp * 50)
    print(f"  {feat:<12}: {imp:.4f}  {barra}")
```

```python
# Cruzamos VUS con CADD y REVEL
df_vus["pos"] = df_vus["pos"].astype(str)
df_vus["chr"] = df_vus["chr"].astype(str)

df_vus_cadd = df_vus.merge(
    df_cadd[["chr", "pos", "ref", "alt", "cadd_phred"]],
    on=["chr", "pos", "ref", "alt"],
    how="left"
)

df_vus_todo = df_vus_cadd.merge(
    df_revel[["grch38_pos", "ref", "alt", "REVEL"]],
    left_on=["pos", "ref", "alt"],
    right_on=["grch38_pos", "ref", "alt"],
    how="left"
).drop(columns=["grch38_pos"])

df_vus_modelo = df_vus_todo[
    df_vus_todo["cadd_phred"].notna() &
    df_vus_todo["REVEL"].notna()
].copy().reset_index(drop=True)

print(f"VUS con CADD y REVEL: {len(df_vus_modelo)}")

# Consultamos gnomAD para las VUS
print(f"Consultando gnomAD...")
afs_vus = []
for i, fila in df_vus_modelo.iterrows():
    af = consultar_gnomad_coords(fila["chr"], fila["pos"], fila["ref"], fila["alt"])
    afs_vus.append(af)
    if (i + 1) % 50 == 0:
        print(f"  {i+1}/{len(df_vus_modelo)} procesadas")
    time.sleep(0.3)

df_vus_modelo["af"] = afs_vus

# Predicción
probs = modelo.predict_proba(df_vus_modelo[FEATURES].values)[:, 1]
df_vus_modelo["prob_patogenica"] = probs

def asignar_prioridad(prob):
    if prob >= 0.7:   return "Alta"
    elif prob >= 0.4: return "Media"
    else:             return "Baja"

df_vus_modelo["prioridad"] = df_vus_modelo["prob_patogenica"].apply(asignar_prioridad)

# Tabla final ordenada (RF-19)
df_resultado = df_vus_modelo[
    ["chr", "pos", "ref", "alt", "cadd_phred", "REVEL", "af", "prob_patogenica", "prioridad"]
].sort_values("prob_patogenica", ascending=False).reset_index(drop=True)

print("\nDistribución de prioridades:")
print(df_resultado["prioridad"].value_counts())
print()
print(df_resultado.head(10))

# Guardamos
df_resultado.to_csv("vus_priorizadas_varai.csv", index=False)
df_modelo.to_csv("brca1_dataset_final.csv", index=False)
print("\nArchivos guardados.")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

# Evaluamos el modelo con validación cruzada para obtener predicciones reales
from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(modelo, X, y, cv=skf)

# Matriz de confusión
cm = confusion_matrix(y, y_pred)
tn, fp, fn, tp = cm.ravel()

print("=" * 50)
print("MATRIZ DE CONFUSIÓN")
print("=" * 50)
print(f"                 Predicho")
print(f"                 Benigna  Patogénica")
print(f"Real Benigna     {tn:>6}  {fp:>10}")
print(f"Real Patogénica  {fn:>6}  {tp:>10}")

print()
print("=" * 50)
print("REPORTE DE CLASIFICACIÓN")
print("=" * 50)
print(classification_report(y, y_pred, target_names=["Benigna", "Patogénica"]))

print()
print("=" * 50)
print("MÉTRICAS INDIVIDUALES")
print("=" * 50)
print(f"Verdaderos Negativos (TN) : {tn}  → Benignas clasificadas correctamente")
print(f"Falsos Positivos     (FP) : {fp}  → Benignas clasificadas como patogénicas")
print(f"Falsos Negativos     (FN) : {fn}  → Patogénicas NO detectadas (crítico)")
print(f"Verdaderos Positivos (TP) : {tp}  → Patogénicas detectadas correctamente")
print()
print(f"Sensibilidad (Recall) clase patogénica: {tp/(tp+fn):.4f}")
print(f"→ El modelo detecta el {tp/(tp+fn)*100:.1f}% de las variantes patogénicas")
print(f"→ Solo el {fn/(tp+fn)*100:.1f}% de las patogénicas pasan desapercibidas")
```
